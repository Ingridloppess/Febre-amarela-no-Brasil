Passo 1: Importar bibliotecas necessárias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

 Passo 2: Carregar o dataset de Meteoros Observados

```python
url = "https://data.nasa.gov/resource/y77d-th95.csv"
data = pd.read_csv(url)
```

Limpeza e Pré-processamento - Filtrar os dados com latitudes e longitudes

```python
data = data.dropna(subset=['reclat', 'reclong'])
data['year'] = pd.to_datetime(data['year'], errors='coerce').dt.year
```

 Passo 3: Explorar os dados


```python
print(data.head())
print(data.info())
```

Passo 4: Visualização de Dados

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Gráfico 1: Distribuição dos meteoros ao longo dos anos

```python
plt.figure(figsize=(14, 7))
sns.histplot(data['year'].dropna(), bins=30, kde=False)
plt.xlabel('Ano')
plt.ylabel('Número de Meteoros Observados')
plt.title('Distribuição dos Meteoros Observados ao Longo dos Anos')
plt.show()

print("\n**Análise da Distribuição dos Meteoros ao Longo dos Anos**\n")
print("No Gráfico 1, observamos a distribuição dos meteoros ao longo dos anos.\n Esta visualização revela tendências e padrões na ocorrência dos meteoros,\n possivelmente destacando períodos de maior atividade ou eficiência nas observações. \n É interessante notar que picos em certos anos podem estar relacionados ao aumento na quantidade de equipamentos de observação ou a eventos astronômicos específicos que desencadearam um número maior de observações.\n")
print("A distribuição temporal é um fator crucial para entender a dinâmica dos meteoros, \n pois permite correlacionar a frequência de eventos com mudanças na tecnologia de observação ou fenômenos celestes específicos.\n À medida que avançamos para o Gráfico 2, iremos explorar a localização geográfica dos meteoros, o que adiciona uma dimensão espacial à nossa análise.\n")
print("--------------------------------------------------------------------------------\n")
```

Gráfico 2: Localização geográfica dos meteoros observados

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['reclong'], data['reclat'], alpha=0.5, s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Localização Geográfica dos Meteoros Observados')
plt.show()

print("\n**Análise da Localização Geográfica dos Meteoros Observados**\n")
print("No Gráfico 2, a localização geográfica dos meteoros é representada em um plano de latitude e longitude. \n Este gráfico permite visualizar onde os meteoros foram observados, \n fornecendo insights sobre áreas de maior incidência ou correlação com regiões específicas do planeta.\n A dispersão dos pontos também pode refletir a densidade de observações em certas regiões, o que pode estar ligado tanto à população quanto à infraestrutura de observação disponível.\n")
print("Agora, passamos para o Gráfico 3, onde analisaremos a distribuição das massas dos meteoros observados.\n")
print("--------------------------------------------------------------------------------\n")
```

Gráfico 3: Distribuição das massas dos meteoros observados

```python
plt.figure(figsize=(14, 7))
sns.histplot(data['mass'].dropna(), bins=30, kde=False)
plt.xlabel('Massa (gramas)')
plt.ylabel('Número de Meteoros')
plt.title('Distribuição das Massas dos Meteoros Observados')
plt.show()

print("\n**Análise da Distribuição das Massas dos Meteoros Observados**\n")
print("O Gráfico 3 apresenta a distribuição das massas dos meteoros observados.\n Este gráfico é fundamental para entender a variabilidade de tamanhos dos meteoros que atingiram a Terra.\n A análise da massa pode ajudar a identificar padrões, como a predominância de meteoros menores ou a presença de alguns eventos com meteoros muito grandes, que são mais raros.\n A distribuição das massas nos dá uma visão clara de quão frequentes são meteoros de diferentes tamanhos.\n")
print("No próximo gráfico, veremos a relação entre a massa dos meteoros e sua localização geográfica.\n")
print("--------------------------------------------------------------------------------\n")
```

Gráfico 4: Massa dos meteoros por localização geográfica

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['reclong'], data['reclat'], s=data['mass']/1000, alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Massa dos Meteoros por Localização Geográfica')
plt.show()

print("\n**Análise da Massa dos Meteoros por Localização Geográfica**\n")
print("O Gráfico 4 combina as informações de massa e localização geográfica dos meteoros. \n As bolhas representam a massa dos meteoros, com tamanhos proporcionais à sua massa,\n e suas posições no gráfico indicam onde caíram na Terra. \n Este gráfico nos permite observar se há uma correlação entre a massa dos meteoros e suas localizações geográficas. Além disso, podemos ver se meteoros de grande massa são mais frequentes em certas regiões ou se estão distribuídos de forma mais uniforme.\n")
print("Agora, exploraremos a correlação entre diferentes variáveis dos meteoros no Gráfico 5.\n")
print("--------------------------------------------------------------------------------\n")
```

Gráfico 5: Matriz de correlação entre variáveis dos meteoros

```python
plt.figure(figsize=(10, 8))
corr_matrix = data[['year', 'mass', 'reclat', 'reclong']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação entre Variáveis dos Meteoros')
plt.show()

print("\n**Análise da Matriz de Correlação entre Variáveis dos Meteoros**\n")
print("No Gráfico 5, temos uma matriz de correlação que revela a relação entre diferentes variáveis dos meteoros, como ano de observação, massa, latitude e longitude.\n A matriz nos permite identificar quais dessas variáveis estão mais relacionadas entre si.\n Por exemplo, uma correlação alta entre a massa e a latitude poderia sugerir um padrão na distribuição geográfica dos meteoros mais massivos. Esta análise ajuda a entender melhor como diferentes características dos meteoros podem estar conectadas.\n")
print("Em seguida, veremos a média anual da massa dos meteoros observados ao longo do tempo.\n")
print("--------------------------------------------------------------------------------\n")
```

Gráfico 6: Média anual da massa dos meteoros observados

```python
mean_mass_per_year = data.groupby('year')['mass'].mean().dropna()
plt.figure(figsize=(14, 7))
plt.plot(mean_mass_per_year.index, mean_mass_per_year.values, marker='o')
plt.xlabel('Ano')
plt.ylabel('Média da Massa (gramas)')
plt.title('Média Anual da Massa dos Meteoros Observados')
plt.show()

print("\n**Análise da Média Anual da Massa dos Meteoros Observados**\n")
print("O Gráfico 6 apresenta a média anual da massa dos meteoros observados. \n Este gráfico nos permite identificar tendências ao longo do tempo, mostrando se os meteoros observados têm se tornado maiores ou menores em média.\n Alterações na média da massa ao longo dos anos podem refletir mudanças nas condições atmosféricas, padrões de observação, ou até mesmo em eventos cósmicos que alteraram o fluxo de meteoros que atingem a Terra.\n")
print("Finalmente, vamos explorar a relação entre a massa dos meteoros e a latitude no próximo gráfico.\n")
print("--------------------------------------------------------------------------------\n")
```

 Gráfico 7: Relação entre massa dos meteoros e latitude

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['reclat'], data['mass'], alpha=0.5, s=10)
plt.xlabel('Latitude')
plt.ylabel('Massa (gramas)')
plt.title('Relação entre Massa dos Meteoros e Latitude')
plt.show()

print("\n**Análise da Relação entre Massa dos Meteoros e Latitude**\n")
print("No Gráfico 7, analisamos a relação entre a massa dos meteoros e a latitude onde foram observados. \n Este gráfico pode indicar se há uma variação significativa nas massas dos meteoros em diferentes latitudes.\n Identificar padrões neste contexto pode ser útil para entender como a geografia da Terra, ou os métodos de observação disponíveis em diferentes regiões, podem influenciar a detecção de meteoros de diferentes tamanhos.\n")
print("Com isso, concluímos a análise visual dos dados de meteoros observados.\n Cada gráfico nos proporcionou insights únicos sobre diferentes aspectos desses eventos fascinantes, desde sua distribuição ao longo do tempo até suas características físicas e localizações geográficas.")
print("--------------------------------------------------------------------------------\n")
```
